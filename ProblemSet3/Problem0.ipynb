```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
SavePlots = True
```

# Problem 0

```python
# Define parameters
dof = 8 #degrees of freedom
low, up = 0, 17 #range

# Define xrange
x_arr = np.linspace(low,up,100)

# Define x2 distribution (pdf)
y_arr = stats.chi2.pdf(x_arr,dof)
```

```python
fig, ax = plt.subplots(figsize=(5,5))

ax.plot(x_arr, y_arr,color='darkred',linewidth=5)

ax.text(14, 0.108, f'DoF={dof}',fontsize=15)
plt.xticks([0.0,5.0,10.0,15.0], fontsize=15)
plt.yticks(fontsize=15)
ax.set_title(r'$\chi^2-Distribution$', fontsize=18)
ax.set_xlabel(r'$\chi^2$', fontsize=15)
ax.set_ylabel('PDF', fontsize=15)

if SavePlots: plt.savefig('Plots/chi2.pdf')

plt.show()
```
